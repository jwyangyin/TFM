<h1>1. Introducción</h1>

<p>
En este cuaderno desarrollamos el <b>Sistema 3</b>: un <b>sistema de recomendación híbrido</b> que combina el
enfoque colaborativo <i>Item-to-Item</i> (Sistema 1) con el enfoque basado en contenido (Sistema 2).
El objetivo es aprovechar lo mejor de ambos sistemas de recomendación: capturar patrones de co-ocurrencia en historiales de usuarios
y, a la vez, incorporar similitud semántica entre productos usando metadatos y texto.
</p>

<p>
Al construir este sistema híbrido buscamos:
</p>
<ul>
  <li><b>Mejorar robustez</b>: si un método falla (p. ej. falta de señal colaborativa), el otro compensa.</li>
  <li><b>Mejorar cobertura</b>: recomendar también productos de la “larga cola”.</li>
  <li><b>Mejorar coherencia</b>: reordenar candidatos colaborativos usando información semántica del contenido.</li>
</ul>
